# 1. Header tags from wikipedia

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup=BeautifulSoup(page.content,"html.parser")
soup
```

# 3.IMDB’s Top rated 100 Indian movies

# 9.Dine Out

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
resturantname =[]
for i in soup.find_all('div', class_="restnt-info cursor"):
    resturantname.append(i.text)
resturantname
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text.split()[0])
rating
```

```python
image=[]
for i in soup.find_all('img',class_="no-img"):
    image.append(i['data-src'])
image
```

```python
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text)
cuisine
```

```python
print(len(resturantname),len(location),len(rating),len(image),len(cuisine))
```

```python
import pandas as pd
data={'resturantname':resturantname,'location':location,'rating':rating,'image':image,}
df=pd.DataFrame(data)
df
```

# 10.publications from Google Scholar

```python
page=requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
rank=[]
for i in soup.find_all('td',class_="gsc_mvt_p"):
    rank.append(i.text)
rank
```

```python
publication=[]
for i in soup.find_all('td',class_="gsc_mvt_t"):
    publication.append(i.text)
publication
```

```python
h5index=[]
for i in soup.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    h5index.append(i.text.split()[0])
h5index
```

```python
h5median=[]
for i in soup.find_all("span",class_="gs_ibl gsc_mp_anchor"):
    h5median.append(i.text)
h5median
```

```python
print(len(rank),len(publication),len(h5index),len(h5median))
```

```python
import pandas as pd
data={'rank':rank,'publication':publication,'h5index':h5index,'h5median':h5median}
df=pd.DataFrame(data)
df
```
